```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.decomposition import PCA 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
housing = pd.DataFrame(pd.read_csv("Housing.csv"))
housing.head()
```

```python
m = len(housing)
m
```

```python
housing.shape
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 
            'hotwaterheating', 'airconditioning', 'parking', 'prefarea', 'price']
df_train = housing[num_vars]
```

```python
df_train.guestroom.replace(('yes', 'no'), (1, 0), inplace=True)
df_train.mainroad.replace(('yes', 'no'), (1, 0), inplace=True)
df_train.basement.replace(('yes', 'no'), (1, 0), inplace=True)
df_train.hotwaterheating.replace(('yes', 'no'), (1, 0), inplace=True)
df_train.airconditioning.replace(('yes', 'no'), (1, 0), inplace=True)
df_train.prefarea.replace(('yes', 'no'), (1, 0), inplace=True)
df_train
```

```python
x0 = df_train.values[:,0] #area
x1 = df_train.values[:,1] #bedrooms
x2 = df_train.values[:,2] #bathrooms
x3 = df_train.values[:,3] #stories
x4 = df_train.values[:,4] #mainroad
x5 = df_train.values[:,5] #guestroom
x6 = df_train.values[:,6] #basement
x7 = df_train.values[:,7] #Hotwater heating
x8 = df_train.values[:,8] #airconditioning
x9 = df_train.values[:,9] #parking
x10 = df_train.values[:,10] #prefarea

y = df_train.values[:,11] #price
```

```python
x0 = x0.reshape(m, 1) 
x1 = x1.reshape(m, 1) 
x2 = x2.reshape(m, 1) 
x3 = x3.reshape(m, 1) 
x4 = x4.reshape(m, 1)
x5 = x5.reshape(m, 1)
x6 = x6.reshape(m, 1)
x7 = x7.reshape(m, 1)
x8 = x8.reshape(m, 1)
x9 = x9.reshape(m, 1)
x10 = x10.reshape(m, 1)

y = y.reshape(m,1)
```

```python
X = np.hstack((x0, x1, x2, x3, x4, x5, x6, x7, x8, x9, x10))
```

```python
#Scaling the data to do the Logistic regression
sc = MinMaxScaler()
x = sc.fit_transform(X)
#Doing the PCA feature extraction for k columns
pca = PCA(n_components=11) 
principalComponents = pca.fit_transform(x) 
principalDf = pd.DataFrame(data = principalComponents)
finalDf = pd.concat([principalDf, housing[['price']]], axis = 1)
x = pd.DataFrame(principalDf.to_numpy())
#Splitting the train and test data
X_train, X_test, Y_train, Y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, 
                                                            random_state=None, shuffle = True)
#Running the Support vector regression machines- linear kernelization
clf = SVR(kernel='linear', C=1e3)
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
#Calculating the metrics- accuracy and mean squared error    
score = clf.score(X_test, Y_test)
msl = mean_squared_error(Y_pred, Y_test)
  
print("For 1 extracted features, accuracy = %.5f and MSE = %.5f" %(score, msl))
```
